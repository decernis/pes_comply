```python
def load_limits(query_data):
    results = {}
    limits = {}
    for i, row in query_data.iterrows():
        results.setdefault(row['COUNTRY'] + '_' + row['FORM_ID'], set())
        limits.setdefault(row['COUNTRY'] + '_' + row['FORM_ID'], set())
    #     if row['COUNTRY'] + '_' + row['FORM_ID'] in results:
        if row['LIMIT'] >= 0:
            limits[row['COUNTRY'] + '_' + row['FORM_ID']].add(row['LIMIT'])
        if row['RESULT_INDICATOR']:
            results[row['COUNTRY'] + '_' + row['FORM_ID']].add(row['RESULT_INDICATOR'])
    #     else:
    #         results[row['COUNTRY'] + '_' + row['FORM_ID']] = set()

    return limits, results
#     print(results)
#         pass
    
```

```python
def generate_rep(sample_id, pes_dic, pes_names, country_ok, pes_ok, country_count, query_data):
    """
    generate matrix, detailed report, summary and write it out
    """
    # generate matrix
    matrix_rows = []
    countries = sorted(set([m for m in country_ok.keys()]))
    for _c in countries:
        m_row = {}
        m_row['country'] = _c
        m_row['country_status'] = country_ok.get(_c, '')
        for _p in pes_dic.keys():
            pes_nam = pes_names.get(_p, '')
            m_row[pes_nam] = pes_ok[_c + '_' + _p]
        matrix_rows.append(m_row)
    matrix_rep = pd.json_normalize(matrix_rows)
    
    # generate summ
    summ_out = {'status':[], 'count':[]}
    for _status in country_count:
        summ_out['status'].append(_status)
        summ_out['count'].append(country_count[_status])
    summ_df = pd.DataFrame.from_dict(summ_out)

#     query_data['PES_NAME'] = query_data['FORM_ID'].map(pes_names)
#     query_data['RES_VALUE'] = query_data['FORM_ID'].map(pes_dic)
    drop_list = ['REQUEST_GUID', 'HIGH_OPERATOR', 'HIGH_LIMIT', 'CITATION_UPDATE_IN_PROGRESS']
    detailed_df = query_data.drop(drop_list, axis=1)
    detailed_df['PES_NAME'] = query_data['FORM_ID'].map(pes_names)
    detailed_df['RES_VALUE'] = query_data['FORM_ID'].map(pes_dic)
    detailed_df = detailed_df.rename(columns={"FORM_ID": "PES_DID"})
    cols = detailed_df.columns.tolist()
    new_cols = cols[-2:] + cols[:-2]
    detailed_df = detailed_df[new_cols]
#     query_data.drop()
    
    writer = pd.ExcelWriter('sam'+sample_id+'_report.xlsx', engine='xlsxwriter')

    # Write each dataframe to a different worksheet.
    matrix_rep.to_excel(writer, sheet_name='CountryReport', index=False)
    detailed_df.to_excel(writer, sheet_name='TabularReport', index=False)
    summ_df.to_excel(writer, sheet_name='Summary', index=False)

    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```
